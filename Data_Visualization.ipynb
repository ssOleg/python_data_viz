```python
from matplotlib.pyplot import figure, autoscale
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('test_housing_data.csv')
```

```python
data.head(40)
```

```python
rooms = data['rooms_per_house']
population = data['population_per_house']
house_value = data['median_house_value']
```

```python
fg, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
fg.tight_layout(pad=5.0)

# Create a figure with 1x2 subplot and make the left subplot active
plt.subplot(1, 3, 1)

# Plot in blue the % of degrees awarded to women in the Physical Sciences
plt.plot(rooms, color='blue')
plt.title('Rooms')

# Make the right subplot active in the current 1x2 subplot grid
plt.subplot(1, 3, 2)

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(population, color='red')
plt.title('Population')

# Make the right subplot active in the current 1x2 subplot grid
plt.subplot(1, 3, 3)

# Plot in red the % of degrees awarded to women in Computer Science
plt.plot(house_value, color='green')
plt.title('House Value')
```

```python
# Add style
# ['Solarize_Light2', '_classic_test_patch', 'bmh', 'classic', 'dark_background', 'fast', 'fivethirtyeight', 
# 'ggplot', 'grayscale', 'seaborn', 'seaborn-bright', 'seaborn-colorblind', 'seaborn-dark', 'seaborn-dark-palette',
# 'seaborn-darkgrid', 'seaborn-deep', 'seaborn-muted', 'seaborn-notebook', 'seaborn-paper', 'seaborn-pastel', 
# 'seaborn-poster', 'seaborn-talk', 'seaborn-ticks', 'seaborn-white', 'seaborn-whitegrid', 'tableau-colorblind10']
plt.style.use('ggplot')

# plot data
plt.plot(rooms, color='blue')

# Add the axis labels
plt.ylabel('Number of rooms')
plt.xlabel('Number of examples')
plt.show()
```

```python
figure(figsize=(18, 10), dpi=100)
autoscale(tight=True)

# plot data
plt.plot(rooms, color='blue', label='Rooms')

# Set legend
plt.legend(loc='upper right')

# Add the axis labels
plt.ylabel('Number of rooms')
plt.xlabel('Number of examples')

# Set custom ticks
plt.xticks([1000,10000,20000], ['1k','10k','20k'])
plt.show()
```

```python
figure(figsize=(18, 10), dpi=100)

# Calculate maximum point values
numbers = range(0, len(rooms))

def find_max(numbers, data):
    data_value = data.max()
    number_value = numbers[data[data == data_value].index[0]]
    return number_value, data_value

number_max, rooms_max = find_max(numbers, rooms)

# plot data
plt.plot(numbers, rooms, color='blue', label='Rooms')

# Set legend
plt.legend(loc='upper right')

# Set the y-axis range
plt.ylim([0, 30])

# Set the x-axis range
plt.xlim([2500, 4500])

# Add annotation
plt.annotate(f'Maximux {round(rooms_max)}', (number_max, rooms_max), (number_max+5, rooms_max+5), arrowprops=dict(facecolor='red'))

# Add the axis labels
plt.ylabel('Number of rooms')
plt.xlabel('Number of examples')

# Set custom ticks
plt.xticks([2500,3500,4500, 5000], ['2.5k','3.5k','4.5k'])
plt.show()
```

```python
figure(figsize=(18, 10), dpi=100)

# plot data
plt.plot(numbers, rooms, color='blue', label='Rooms')

# Set legend
plt.legend(loc='upper right', fontsize=12)

# Add annotation
plt.annotate(f'Maximux {round(rooms_max)}', (number_max, rooms_max), (number_max+5, rooms_max+5), arrowprops=dict(facecolor='red'))

# Add the axis labels
plt.ylabel('Number of rooms')
plt.xlabel('Number of examples')

#Add the title
plt.title("Number of rooms per house")

# Set both x-axis and y-axis range
plt.axis([2500, 4500, 0, 30])

# Set custom ticks
plt.xticks([2500,3500,4500, 5000], ['2.5k','3.5k','4.5k'])

plt.show()
```

```python
# Calculate maximum values for population
pnumber_max, population_max = find_max(numbers, population)

figure(figsize=(18, 10), dpi=100)

# plot data
plt.plot(numbers, rooms, color='blue', label='Rooms')

# plot data
plt.plot(numbers, population, color='green', label='Population')

# Set legend
plt.legend(loc='upper right', fontsize=22)

# Add annotation
plt.annotate(f'Maximux {round(rooms_max)}', (number_max, rooms_max), (number_max+5, rooms_max+5), arrowprops=dict(facecolor='red'))
plt.annotate(f'Maximux {round(population_max)}', (pnumber_max, population_max), (pnumber_max+10, population_max+10), arrowprops=dict(facecolor='red'))


# Add the axis labels
plt.ylabel('Number of rooms')
plt.xlabel('Number of examples')

#Add the title
plt.title("Number of rooms per house")

# Set the y-axis range
plt.ylim([0, 30])


plt.show()
```
